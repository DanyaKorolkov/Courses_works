```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 101)
y = x

u = [1]*101

a = 2 * x
```

```python
plt.plot(x, y, label = 'liklehood', c = 'blue')

plt.plot(x, a, c = 'red', label = 'posterior')
plt.axvline(x = 1, ymax=0.95, ymin=0, c = 'red')

plt.plot(x, u, label = 'prior',       c = 'black', linestyle = 'dashed')
plt.axvline(x = 0, ymax=0.5, ymin=0, c = 'black', linestyle = 'dashed')
plt.axvline(x = 1, ymax=0.5, ymin=0, c = 'black', linestyle = 'dashed')

plt.xlabel('$\\theta$')

plt.legend()
plt.grid()
plt.show()
```
