```python
import numpy as np
import matplotlib.pyplot as plt
```

Первая итерация

```python
x = np.linspace(0, 1, 101)
y = x

u = [1]*101
```

```python
plt.plot(x, y, label = 'liklehood', c = 'blue')

plt.plot(x, 2*x, c = 'red', label = 'posterior')
plt.axvline(x = 1, ymax=0.95, ymin=0, c = 'red')

plt.plot(x, u, label = 'prior',       c = 'black', linestyle = 'dashed')
plt.axvline(x = 0, ymax=0.5, ymin=0, c = 'black', linestyle = 'dashed')
plt.axvline(x = 1, ymax=0.5, ymin=0, c = 'black', linestyle = 'dashed')

plt.xlabel('$\\theta$')

plt.legend()
plt.grid()
plt.show()
```

Примеры Бета-распределений.

```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
a, b = 1, 1
mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
```

```python
x = np.linspace( 0, 1, 101)

plt.plot(x, beta.pdf(x, 1, 1), '--', label='$\\alpha = 1, \\beta = 1$')
plt.plot(x, beta.pdf(x, 1, 2), '--', label='$\\alpha = 1, \\beta = 2$')
plt.plot(x, beta.pdf(x, 1, 3), '--', label='$\\alpha = 1, \\beta = 3$')
plt.plot(x, beta.pdf(x, 2, 1), label='$\\alpha = 2, \\beta = 1$')
plt.plot(x, beta.pdf(x, 2, 2), label='$\\alpha = 2, \\beta = 2$')
plt.plot(x, beta.pdf(x, 2, 3), label='$\\alpha = 2, \\beta = 3$')

plt.grid()
plt.legend()
plt.show()
```

```python
x = np.linspace( 0, 1, 101)

plt.plot(x, beta.pdf(x, 0.5, 0.5), '--', label='$\\alpha = 0.5, \\beta = 0.5$')
plt.plot(x, beta.pdf(x, 1, 0.5),   '--', label='$\\alpha = 1, \\beta = 0.5$')
plt.plot(x, beta.pdf(x, 3, 3),  label='$\\alpha = 3, \\beta = 3$')
plt.plot(x, beta.pdf(x, 4, 4),  label='$\\alpha = 4, \\beta = 4$')
plt.plot(x, beta.pdf(x, 5, 5),  label='$\\alpha = 5, \\beta = 5$')
plt.plot(x, beta.pdf(x, 10, 5), label='$\\alpha = 10, \\beta = 5$')

plt.grid()
plt.legend()
plt.show()
```
